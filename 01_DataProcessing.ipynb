### Initial Cleanup

```python
data['CRASH DATETIME'] = pd.to_datetime(data['CRASH DATE'] + ' ' + data['CRASH TIME'], errors='coerce')
```

```python
data.drop(['CRASH DATE', 'CRASH TIME'], axis=1, inplace=True)
```

```python
# Drop duplicate rows if any
data.drop_duplicates(inplace=True)
```

### Handling Missing Values

```python
# Check missing values
missing = data.isnull().sum().sort_values(ascending=False)
```

```python
missing
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.isnull(), cbar=False, cmap='coolwarm')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Drop rows with missing target columns or critical location data
data.dropna(subset=['LATITUDE', 'LONGITUDE', 'CRASH DATETIME'], inplace=True)
```

```python
# Remaining: drop columns with too many nulls (e.g., if >90% are null)
null_threshold = 0.9
to_drop = data.columns[data.isnull().mean() > null_threshold]
data.drop(columns=to_drop, inplace=True)
```

```python
# Fill missing boroughs based on ZIP CODE if possible
if 'ZIP CODE' in data.columns and 'BOROUGH' in data.columns:
    zip_borough_map = data.dropna(subset=['ZIP CODE', 'BOROUGH']).drop_duplicates(subset=['ZIP CODE'])[['ZIP CODE', 'BOROUGH']]
    zip_borough_dict = dict(zip(zip_borough_map['ZIP CODE'], zip_borough_map['BOROUGH']))
    data['BOROUGH'] = data['BOROUGH'].fillna(data['ZIP CODE'].map(zip_borough_dict))
```

```python
# For categorical features: fill with 'Unspecified'
cat_cols = [col for col in data.columns if data[col].dtype == 'object']
data[cat_cols] = data[cat_cols].fillna('Unspecified')
```

```python
# For numerical features: fill with 0
num_cols = [col for col in data.columns if data[col].dtype in ['int64', 'float64']]
data[num_cols] = data[num_cols].fillna(0)
```

```python
# Lowercase column names for consistency
data.columns = [col.lower().replace(' ', '_') for col in data.columns]
```

```python
data['crash_datetime'] = pd.to_datetime(data['crash_datetime'], errors='coerce')
```

```python
print("Data shape after cleaning:", data.shape)
print("Remaining missing values:", data.isnull().sum().sum())
```
